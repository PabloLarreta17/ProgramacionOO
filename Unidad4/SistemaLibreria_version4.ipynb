```python
from datetime import datetime
from abc import ABC, abstractmethod
class Persona(ABC):
  def __init__(self, nombre, correo, fecha_nacimiento):
    self.nombre = nombre
    self.correo = correo
    self.fecha_nacimiento = fecha_nacimiento

  @abstractmethod
  def mostrar_info(self):
    "Metodo abstracto: las subclases deben redefinirlo"
    pass

class Registrable(ABC):
  @abstractmethod
  def registrar_actividade(self,accion):
    pass

  class Estudiante(Persona):
    def __init__(self,num_control,nombre,carrera,fecha_nacimiento,telefono,correo):
      super().__init__(nombre,correo,fecha_nacimiento)
      self.num_control = num_control
      self.carrera = carrera
      self.telefono = telefono

    def mostrar_info(self):
      return(f"[Estudiante] Nombre: {self.nombre}|"
            f"Numero Control: {self.num_control}"|
             f"Carrera: {self.carrera}|Telefono: {self.telefono}")

class Trabajador(Persona,Registrable):
  def __init__(self,num_trabajador,nombre,correo,fecha_nacimiento,departamento,usuario,contraseña):
    super().__init__(nombre,correo,fecha_nacimiento)
    self.num_trabajador = num_trabajador
    self.departamento = departamento
    self.usuario = usuario
    self.contrasena = contraseña
    self.actividades = []

  def mostrar_info(self):
    return(f"[Trabajador]Usuario:{self.usuario}, Num trabajador:{self.num_trabajador},Nombre:{self.nombre}")

  def registrar_actividad(self.accion)
  fecha_actual= datetime.now().strftime("%Y-%m-%d %H:%M:%S")
  registro = f"El {fecha_actual}={self.nombre} realizo {accion}"
  self.actividades.append(registro)

class Producto:
    def __init__(self, codigo, nombre, categoria, marca, descripcion, precio, stock):
        self.codigo = codigo
        self.nombre = nombre
        self.categoria = categoria
        self.marca = marca
        self.descripcion = descripcion
        self.precio = precio
        self.stock = stock

    def mostrar_info(self):
        print(f"[{self.codigo}] {self.nombre} - {self.marca} | {self.categoria} | "
              f"${self.precio} | Stock: {self.stock} | {self.descripcion}")


class Cliente:
    def __init__(self, id_cliente, nombre, telefono, correo):
        self.id_cliente = id_cliente
        self.nombre = nombre
        self.telefono = telefono
        self.correo = correo

    def mostrar_info(self):
        print(f"[{self.id_cliente}] {self.nombre} | Tel: {self.telefono} | Correo: {self.correo}")


class Venta:
    contador_ventas = 1

    def __init__(self, codigo_producto, id_cliente, fecha, cantidad, total):
        self.folio = Venta.contador_ventas
        Venta.contador_ventas += 1
        self.codigo_producto = codigo_producto
        self.id_cliente = id_cliente
        self.fecha = fecha
        self.cantidad = cantidad
        self.total = total

    def mostrar_info(self):
        print(f"Folio: {self.folio} | Producto: {self.codigo_producto} | "
              f"Cliente: {self.id_cliente} | Fecha: {self.fecha} | Cantidad: {self.cantidad} | Total: ${self.total}")


# --- Funciones del sistema ---

def alta_producto():
    print("\n--- Alta de Producto ---")
    codigo = input("Código del producto: ")
    for prod in productos:
        if prod.codigo == codigo:
            print("¡Ya existe un producto con ese código!")
            return
    nombre = input("Nombre: ")
    categoria = input("Categoría: ")
    marca = input("Marca: ")
    descripcion = input("Descripción: ")
    try:
        precio = float(input("Precio: "))
        stock = int(input("Cantidad en stock: "))
    except ValueError:
        print("Datos numéricos inválidos.")
        return
    nuevo_producto = Producto(codigo, nombre, categoria, marca, descripcion, precio, stock)
    productos.append(nuevo_producto)
    print("Producto registrado exitosamente.")


def alta_cliente():
    print("\n--- Alta de Cliente ---")
    id_cliente = input("ID del cliente: ")
    for c in clientes:
        if c.id_cliente == id_cliente:
            print("¡Ya existe un cliente con ese ID!")
            return
    nombre = input("Nombre: ")
    telefono = input("Teléfono: ")
    correo = input("Correo electrónico: ")
    nuevo_cliente = Cliente(id_cliente, nombre, telefono, correo)
    clientes.append(nuevo_cliente)
    print("Cliente registrado exitosamente.")

def alta_trabajador():
    print("\n--- Alta de Trabajador ---")
    num_trabajador = input("Número de trabajador: ")
    for t in trabajadores:
        if t.num_trabajador == num_trabajador:
            print("¡Ya existe un trabajador con ese número!")
            return
    nombre = input("Nombre: ")
    correo = input("Correo electrónico: ")
    fecha_nacimiento = input("Fecha de nacimiento (dd/mm/aaaa): ")
    departamento = input("Departamento: ")
    usuario = input("Usuario de Acceso: ")
    contrasena = input("Contraseña: ")

    nuevo_trabajador = Trabajador(num_trabajador, nombre, correo, fecha_nacimiento, departamento, usuario, contrasena)
    trabajadores.append(nuevo_trabajador)
    print(f"Trabajador {nombre} registrado exitosamente.")

def mostrar_trabajadores():
    print("\n--- Lista de Trabajadores ---")
    if not trabajadores:
        print("No hay trabajadores registrados.")
    else:
        for t in trabajadores:
            t.mostrar_info()

def inicio_sesion():
    print("\n--- Inicio de Sesión ---")
    usuario = input("Usuario: ")
    contrasena = input("Contraseña: ")
    for t in trabajadores:
        if t.usuario == usuario and t.contrasena == contrasena:
            print(f"Inicio de sesión exitoso. Bienvenido, {t.nombre}")
            return True
    print("Usuario o contraseña incorrectos.")
    return False



def consultar_producto():
    print("\n--- Consultar Producto por Código ---")
    codigo = input("Ingrese el código del producto: ")
    for prod in productos:
        if prod.codigo == codigo:
            prod.mostrar_info()
            return
    print("Producto no encontrado.")


def consultar_cliente():
    print("\n--- Consultar Cliente por ID ---")
    id_cliente = input("Ingrese el ID del cliente: ")
    for c in clientes:
        if c.id_cliente == id_cliente:
            c.mostrar_info()
            return
    print("Cliente no encontrado.")


def mostrar_todos_productos():
    print("\n--- Lista de Todos los Productos ---")
    if not productos:
        print("No hay productos registrados.")
    else:
        for prod in productos:
            prod.mostrar_info()


def mostrar_todos_clientes():
    print("\n--- Lista de Todos los Clientes ---")
    if not clientes:
        print("No hay clientes registrados.")
    else:
        for c in clientes:
            c.mostrar_info()


def realizar_venta():
    print("\n--- Registrar Venta ---")
    codigo = input("Código del producto: ")
    prod = next((p for p in productos if p.codigo == codigo), None)
    if not prod:
        print("Producto no encontrado.")
        return
    try:
        cantidad = int(input("Cantidad a vender: "))
    except ValueError:
        print("Cantidad inválida.")
        return
    if cantidad > prod.stock:
        print("No hay suficiente stock disponible.")
        return
    id_cliente = input("ID del cliente (999 si es público en general): ")
    if id_cliente != "999":
        cliente = next((c for c in clientes if c.id_cliente == id_cliente), None)
        if not cliente:
            print("Cliente no encontrado.")
            return
    fecha = input("Fecha de la venta (dd/mm/aaaa): ")
    total = prod.precio * cantidad
    venta = Venta(codigo, id_cliente, fecha, cantidad, total)
    ventas.append(venta)
    prod.stock -= cantidad
    print(f"Venta registrada correctamente. Folio: {venta.folio}")


def mostrar_historial_ventas():
    print("\n--- Historial de Ventas ---")
    if not ventas:
        print("No hay ventas registradas.")
    else:
        for v in ventas:
            v.mostrar_info()


# --- Menú principal ---

def mostrar_menu():
    print("\n--- SISTEMA DE FERRETERÍA ---")
    print("1. Dar de alta producto")
    print("2. Dar de alta cliente")
    print("3. Consultar producto por código")
    print("4. Consultar cliente por ID")
    print("5. Mostrar todos los productos")
    print("6. Mostrar todos los clientes")
    print("7. Registrar venta")
    print("8. Mostrar historial de ventas")
    print("9. Dar de alta trabajador")
    print("10. Mostrar todos los trabajadores")
    print("11. Salir del sistema")


def iniciar_sistema():
    trabajador_Activo = None
    while True:
        if not trabajador_Activo:
            trabajador_Activo = inicio_sesion()
            if not trabajador_Activo:
                continuar = input("¿Desea intentar iniciar sesión de nuevo? (s/n): ").lower()
                if continuar != 's':
                    print("Saliendo del sistema. ¡Hasta luego!")
                    break
                else:
                    continue
        mostrar_menu()
        opcion = input("Seleccione una opción: ")
        if opcion == "1":
            alta_producto()
        elif opcion == "2":
            alta_cliente()
        elif opcion == "3":
            consultar_producto()
        elif opcion == "4":
            consultar_cliente()
        elif opcion == "5":
            mostrar_todos_productos()
        elif opcion == "6":
            mostrar_todos_clientes()
        elif opcion == "7":
            realizar_venta()
        elif opcion == "8":
            mostrar_historial_ventas()
        elif opcion == "9":
            alta_trabajador()
        elif opcion == "10":
            mostrar_trabajadores()
        elif opcion == "11":
            print("Cierre de Sesion")
            trabajador_Activo = None
        elif opcion == "12":
            print("Saliendo del sistema. ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Intente de nuevo.")


# --- Programa principal ---
productos = []
clientes = []
ventas = []
trabajadores = []

admin = Trabajador("0001", "Admininistrador", "admin@gmail.com", "01/01/1990", "Sistemas", "admin", "admin")
admin1 = Trabajador("0002", "Admininistrador", "freddy@gmail.com", "19/06/2006", "Sistemas", "freddy", "admin")
trabajadores.append(admin)
trabajadores.append(admin1)

# Iniciar sistema
iniciar_sistema()
```
